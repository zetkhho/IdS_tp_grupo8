En esta fase se construirá el conjunto final de datos que será utilizado en el análisis y el modelado. Para ello, se realizará una copia de trabajo del dataset original, se eliminarán registros duplicados, se tratarán valores faltantes en variables relevantes para el estudio y se filtrarán transacciones no representativas de ventas efectivas, como aquellas con cantidades o precios no positivos. Además, se generarán variables derivadas que permitan enriquecer el análisis temporal y económico de las ventas.

```python
# copia del dataset.
df_prep = df.copy()

print ("tamaño inicial: ", df_prep.shape)
df_prep.head(20)
```

```python
# eliminacion de duplicados
print ("Duplicados antes: ", df_prep.duplicated().sum())

df_prep.drop_duplicates(inplace=True)
print ("Duplicados despues: ", df_prep.duplicated().sum())

print ("Tamaño del dataset despues de la eliminacion de duplicados: ", df_prep.shape)
```

```python
# eliminacion de filas sin descripcion del producto
print ("Registros sin descripcion: ", df_prep["Description"].isnull().sum())
df_prep.dropna(subset=["Description"], inplace=True)
print ("Registros sin descripcion: ", df_prep["Description"].isnull().sum())
print ("Tamaño del dataset luego de la eliminación: ", df_prep.shape)
```

```python
print ("Nulos en CustomerID: ",df_prep["CustomerID"].isnull().sum())
```

```python
# filtrar cantidad y precios negativos.
print ("Registros con Quantity <= 0: ",(df_prep['Quantity']<=0).sum())
print ("Registros con UnitPrice <= 0: ",(df_prep['UnitPrice']<=0).sum())

df_prep = df_prep[(df_prep['Quantity'] > 0) & (df_prep['UnitPrice'] > 0)]

print ("Tamaño del dataset despues del filtro: ",df_prep.shape)
```

```python
# variables para posterior analisis.
# variable de monto total
df_prep["TotalAmount"] = df_prep["Quantity"] * df_prep["UnitPrice"]

#variables temporales
df_prep["Year"] = df_prep["InvoiceDate"].dt.year
df_prep["Month"] = df_prep["InvoiceDate"].dt.month
df_prep["Day"] = df_prep["InvoiceDate"].dt.day
df_prep["YearMonth"]= df_prep["InvoiceDate"].dt.to_period('M').astype(str)
df_prep.head(10)
```

Aunque en la limpieza inicial se eliminaron duplicados, registros sin descripción relevante y transacciones con cantidades o precios no positivos, todavía se observan valores extremos en variables como Quantity, UnitPrice y TotalAmount. Por este motivo, se realizará un análisis específico de outliers utilizando el criterio del rango intercuartílico (IQR), no con el objetivo de eliminarlos automáticamente, sino para determinar si corresponden a errores, casos atípicos válidos o valores que deban tratarse de forma especial en etapas posteriores.

```python
columnas_out = ["Quantity", "UnitPrice", "TotalAmount"]

for columna in columnas_out:
    Q1 = df_prep[columna].quantile(0.25)
    Q3 = df_prep[columna].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    cantidad_outliers = (
        (df_prep[columna] < limite_inferior) |
        (df_prep[columna] > limite_superior)
    ).sum()

    print(f"Variable: {columna}")
    print(f"Q1: {Q1}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Límite inferior: {limite_inferior}")
    print(f"Límite superior: {limite_superior}")
    print(f"Cantidad de outliers: {cantidad_outliers}")
    print("-" * 50)
```

```python
for columna in ["Quantity", "UnitPrice", "TotalAmount"]:
    Q1 = df_prep[columna].quantile(0.25)
    Q3 = df_prep[columna].quantile(0.75)
    IQR = Q3 - Q1

    li = Q1 - 1.5 * IQR
    ls = Q3 + 1.5 * IQR

    outliers = ((df_prep[columna] < li) | (df_prep[columna] > ls)).sum()
    porcentaje = outliers / len(df_prep) * 100

    print(f"{columna}: {outliers} outliers ({porcentaje:.2f}%)")
```

Los valores atípicos fueron detectados y analizados, pero no se eliminaron automáticamente, ya que una parte de ellos podría corresponder a transacciones reales, compras excepcionales o registros comerciales válidos. Por este motivo, se conservaron en la base preparada, aunque se utilizaron recortes visuales mediante percentiles para facilitar la interpretación gráfica.

### **Objetivo de modelado**

En esta fase se definió como variable objetivo el monto total vendido por mes, calculado a partir de la suma de "TotalAmount" para cada período "YearMonth". Esta decisión se vincula directamente con el objetivo del estudio, que busca anticipar ventas futuras a partir del comportamiento histórico.

Como técnica inicial de modelado se seleccionó Regresión Lineal, debido a su simplicidad, interpretabilidad y utilidad como modelo base para analizar tendencias temporales en las ventas mensuales.

```python
#Acá debés cargar las librerías con las que trabajarás en esta fase
# import de las librerias que vamos a usar
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

### **Construcción de la base mensual de modelado**

```python
# construcción de la serie mensual del modelado
ventas_mensuales = df_model.groupby("YearMonth")["TotalAmount"].sum().reset_index()

ventas_mensuales["YearMonth"] = pd.to_datetime(ventas_mensuales["YearMonth"])
ventas_mensuales = ventas_mensuales.sort_values("YearMonth").reset_index(drop=True)

# Crear indice temporal numerico (se usa en la regresion lineal, ya que no entiende fechas pero si número)
ventas_mensuales["TimeIndex"] = np.arange(len(ventas_mensuales))

# ahora tenemos meses y montos totales vendidos.
ventas_mensuales.head(10)
```

### Visualización de la serie antes de modelar.

```python
# Visualizamos la base que creamos.
plt.figure(figsize=(6,3))
plt.plot(ventas_mensuales["YearMonth"], ventas_mensuales["TotalAmount"], marker = "o")
plt.title("Serie temporal de ventas mensuales")
plt.xlabel("Fecha")
plt.ylabel("Monto Total")
plt.xticks(rotation=45)
plt.show()
```

```python
# Definir X e Y
X = ventas_mensuales[["TimeIndex"]] # X va a ser TimeIndex como entrada
y = ventas_mensuales["TotalAmount"] # Y va a ser el TotalAmount mensual como salida

print ("X:")
print (X.head(10))
print ("\ny:")
print (y.head(10))
```

### Entrenar el modelo y ver resultados

```python
# Entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, y)
predicciones = modelo.predict(X)

# ver resultados
ventas_mensuales["Predicciones"] = predicciones.round(2)
ventas_mensuales.head(10)
```

### Graficar valores reales Vs Predichos

```python
# Grafica de valores reales vs las predicciones (para ver si los valores se comportan parecido, los predichos vs los reales, sin ser aún futuros solo históricos)
plt. figure(figsize=(6,3))
plt.plot(ventas_mensuales['YearMonth'],ventas_mensuales['TotalAmount'], marker='o', label='Valores reales')
plt.plot(ventas_mensuales['YearMonth'],ventas_mensuales['Predicciones'], marker='o', label='Predicciones')
plt.title('Valores reales vs predicciones')
plt.xlabel('Fecha')
plt.ylabel('Monto total')
plt.legend()
plt.show()
```

Se entrenó un modelo de Regresión Lineal utilizando como variable de entrada el índice temporal (TimeIndex) y como variable objetivo el monto total vendido por mes (TotalAmount).

Las predicciones generadas por el modelo aparecen con varios decimales porque se calculan como valores continuos de tipo float, aunque para fines de presentación pueden redondearse.

En la comparación entre valores reales y predichos se observa que el modelo ajusta una línea recta creciente, ya que la Regresión Lineal simple representa únicamente una tendencia global en el tiempo. Por este motivo, logra captar la dirección general de crecimiento de las ventas, pero no reproduce adecuadamente las variaciones mensuales, los picos pronunciados ni las caídas bruscas presentes en la serie.

### Prediccion futura simple

```python
futuros = pd.DataFrame({"TimeIndex":np.arange(len(ventas_mensuales), len(ventas_mensuales)+12)})
pred_futuras = modelo.predict(futuros)
pred_futuras = pred_futuras.round(2)
print ("Predicciones para los próximos 12 meses:")
print (pred_futuras)
```

### Armar tabla con meses futuros


```python
ultimo_mes = ventas_mensuales["YearMonth"].max()
meses_futuros = pd.date_range(start=ultimo_mes + pd.offsets.MonthBegin(1), periods=12, freq="MS")
df_futuro = pd.DataFrame({"YearMonth":meses_futuros,"Predicciones":pred_futuras})
df_futuro["Predicciones"] = df_futuro["Predicciones"].round(2)
print ("Predicciones para los próximos 12 meses:")
print (df_futuro)
```

### Graficos historicos + futuros

```python
from matplotlib.lines import lineStyles
# Visualizaciones
plt.figure(figsize=(6,3))
plt.plot(ventas_mensuales["YearMonth"],ventas_mensuales["TotalAmount"], marker="o", label="Valores reales")
plt.plot(ventas_mensuales["YearMonth"],ventas_mensuales["Predicciones"], marker="o", label="Predicciones")
plt.plot(df_futuro["YearMonth"], df_futuro["Predicciones"], marker="o", linestyle="--", label="Predicción futura")

plt.title("Serie temporal de ventas mensuales")
plt.xlabel("Fecha")
plt.ylabel("Monto Total")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

El modelo de Regresión Lineal permitió construir una primera aproximación predictiva sobre las ventas mensuales. Los resultados muestran que logra captar la tendencia general creciente de la serie, pero no reproduce adecuadamente las variaciones más marcadas, como picos y caídas bruscas. Esto se refleja tanto en la visualización como en el valor de R², que indica una capacidad explicativa limitada. Por este motivo, la Regresión Lineal se considera un modelo base adecuado, aunque no necesariamente el más óptimo para representar el comportamiento completo de la serie temporal.

### Modelo alternativo: Regresión Polinómica

Como modelo alternativo se implementara una Regresión Polinómica, con el objetivo de capturar mejor posibles curvaturas en la evolución temporal de las ventas mensuales. A diferencia de la Regresión Lineal simple, este enfoque incorpora términos de orden superior del índice temporal, permitiendo un ajuste más flexible sobre la serie histórica. Esto resulta útil cuando los datos presentan cambios no lineales que una recta no logra representar adecuadamente.

```python
# Regresión Polinómica
grado = 3

poly = PolynomialFeatures(degree=grado, include_bias=False)
X_poly = poly.fit_transform(X)

modelo_poly = LinearRegression()
modelo_poly.fit(X_poly, y)

predicciones_poly = modelo_poly.predict(X_poly)

ventas_mensuales["Predicciones_Polinomicas"] = predicciones_poly.round(2)

ventas_mensuales.head(10)
```

### Graficas real vs lineal vs polinómica

```python
plt.figure(figsize=(6,3))
plt.plot(ventas_mensuales["YearMonth"], ventas_mensuales["TotalAmount"], marker="o", label="Valores reales")
plt.plot(ventas_mensuales["YearMonth"], ventas_mensuales["Predicciones"], marker="o", label="Predicciones Lineal")
plt.plot(ventas_mensuales["YearMonth"], ventas_mensuales["Predicciones_Polinomicas"], marker="o", label="Predicciones Polinómica")

plt.title("Serie temporal de ventas mensuales")
plt.xlabel("Fecha")
plt.ylabel("Monto Total")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

En esta fase se construyeron dos modelos predictivos para estimar el monto total vendido por mes a partir del índice temporal de la serie. En primer lugar, se implementó una Regresión Lineal como modelo base, por su simplicidad e interpretabilidad. Sin embargo, este modelo mostró limitaciones para representar las variaciones de la serie temporal, ya que solo captó una tendencia general creciente.

Luego, se implementó una Regresión Polinómica como modelo alternativo. Se probaron distintos grados y se observó que el modelo de grado 3 ofreció un mejor ajuste que la regresión lineal y que la polinómica de grado 2.

Aunque al aumentar el grado del polinomio el modelo mejora, no necesariamente esto implica un mejor modelo predictivo. Grados muy altos pueden generar sobreajuste, representando también el ruido o fluctuaciones particulares del conjunto histórico. Dado que la serie mensual disponible es reducida, se optó por un grado 3 como compromiso entre ajuste, simplicidad e interpretabilidad.

Por lo tanto, para este trabajo se considera que la Regresión Polinómica de grado 3 representa mejor el comportamiento histórico de las ventas mensuales y resulta más adecuada como modelo predictivo principal para continuar con la evaluación.

**Evaluación del modelo**

En esta fase se evalúa el rendimiento del modelo predictivo seleccionado, con el objetivo de determinar qué tan bien representa el comportamiento histórico de las ventas mensuales. Para ello se analizarán métricas de error y de ajuste, comparando especialmente el desempeño de la Regresión Lineal y de la Regresión Polinómica de grado 3. La finalidad es justificar cuál de los modelos resulta más adecuado para este estudio.

### **Métricas iniciales**

```python
# Métricas del modelo lineal
mae = mean_absolute_error(y, predicciones) # Mide en promedio, cuánto me estoy equivocando
mse = mean_squared_error(y, predicciones)  # Mide el mae pero al cuadrado, que tan grande son los errores, da mas importancia a los errores mas grandes.
rmse = np.sqrt(mse) # vuelve a la escala anterior al mae pero dando mas importancia a los errores mas grandes
r2 = r2_score(y, predicciones) # Mide que porcentaje de la variabilidad del dato explica tu modelo (mas cercano a 1 mejor, negativo es mejor adivinar)

print ("Métricas del modelo:")
print ("MAE:", round(mae,2))
print ("MSE:", round(mse,2))
print ("RMSE:", round(rmse,2))
print ("R2:", round(r2,4))

# Métricas del modelo polinómico
mae_poly = mean_absolute_error(y, predicciones_poly)
mse_poly = mean_squared_error(y, predicciones_poly)
rmse_poly = np.sqrt(mse_poly)
r2_poly = r2_score(y, predicciones_poly)

print ("\nMétricas del modelo polinómico:")
print ("MAE:", round(mae_poly,2))
print ("MSE:", round(mse_poly,2))
print ("RMSE:", round(rmse_poly,2))
print ("R2:", round(r2_poly,4))
```

### tabla comparativa

```python
metricas_modelos = pd.DataFrame({
    "Modelo": ["Regresión Lineal", "Regresión Polinómica grado 3"],
    "MAE": [round(mae, 2), round(mae_poly, 2)],
    "MSE": [round(mse, 2), round(mse_poly, 2)],
    "RMSE": [round(rmse, 2), round(rmse_poly, 2)],
    "R2": [round(r2, 4), round(r2_poly, 4)]
})

metricas_modelos
```

**Interpretación de resultados**

La comparación de métricas muestra que la Regresión Polinómica de grado 3 supera al modelo de Regresión Lineal en todas las medidas consideradas. Presenta menores valores de MAE, MSE y RMSE, lo que indica un menor error de predicción, y un coeficiente R² más alto, lo que refleja una mejor capacidad para explicar la variabilidad de las ventas mensuales.

En particular, mientras la Regresión Lineal solo logra captar una tendencia creciente general, la Regresión Polinómica representa mejor las variaciones de la serie histórica. Aun así, el valor de R² obtenido no es cercano a 1, por lo que el modelo todavía tiene limitaciones y no debe interpretarse como una predicción exacta, sino como una aproximación útil al comportamiento de las ventas.

La matriz de confusión no se aplica en este trabajo, ya que el problema abordado es de regresión y no de clasificación. **Conclusión**

A partir de las métricas evaluadas, se concluye que la Regresión Polinómica de grado 3 presenta un mejor desempeño que la Regresión Lineal para representar el comportamiento histórico de las ventas mensuales.
